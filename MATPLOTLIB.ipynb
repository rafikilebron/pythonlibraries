## GRAFICAR DATOS

Si a la función `plot()` sólo se le da una matriz de datos, `la usará como coordenadas en el eje vertical`, y utilizará el `índice` de cada dato como la `coordenada horizontal`.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
# show() muestra la gráfica
plt.show()
```

También podemos graficar coordenadas **_(x,y)_** introduciendo como parámetros los `puntos X` y los `puntos Y`

* **NOTA**: la cantidad de números de la lista X e Y deben ser iguales.

```python
x = [1, 2, 5, 8, 7]
y = [5, 7, 6, 4, 0]

# Primer parámetro PTOS. X, segundo parámetro PTOS. Y
plt.plot(x,y)
plt.show()
```

### Ajustar límites de los ejes

##### .axis( [x_min, x_max, y_min, y_max] )

Esta función permite ajustar los límites del eje X y del eje Y

* **NOTA:** no cambia el tamaño del gráfico.

```python
plt.plot(x,y)
plt.axis([-4, 8, 0, 8]) # [xmin, xmax, ymin, ymax]
plt.show()
```

### Graficar Funciones Matemáticas

El siguiente segmento de código grafica una función cuadrada, lo que corresponde a una parábola

```python
import numpy as np

# Crear un arreglo unidimensional de 100 puntos equidistantes de -2 a 2
x = np.linspace(-2, 2, 200)
y = x**2

plt.plot(x,y)
plt.show()
```

También es posible con la función seno

```python
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

### Modificar estilo de la gráfica

Es posible activar la `cuadrícula`, el `título` de la gráfica, las `etiquetas` de cada eje, así como el `tamaño de fuente` de estas.

```python
plt.plot(x,x**2)

# Título
plt.title("Parábola")
# Etiquetas de los ejes
plt.xlabel("x", fontsize=14)
plt.ylabel("y = x**2",fontsize=14)
# Mostrar cuadrícula
plt.grid(True)

plt.show()
```

### Superponer gráficas

Si en los parámetros de `plot()` introducimos 4 listas de puntos, obtenemos dos gráficas en la misma cuadrícula

```python
x = np.linspace(-2, 2, 100)
y = x**2
y1 = x+1

plt.plot(x, y, x, y1)
plt.show()
```

##### Cambiar `estilo` de las líneas y añadir `leyendas`

Dentro del parámetro de `plot()` podemos introducir el:
* **Tipo de línea:** contínua o punteada.
* **Color de línea:** verde, roja, azúl, etc.

La sintaxis es la siguiente:

```python
# b-- línea azúl discontinua
# g línea verde contínua
plt.plot(x, y, 'b--', x, y1, 'g')
plt.show()
```

Una forma más amigable de escribirlo sería la siguiente:

```python
plt.plot(x, y, 'b--')
plt.plot(x, y1, 'g')
plt.show()
```

Podemos diferenciar las funciones añadiendo una `leyenda`, esto lo introducimos como un parámetro adicional en el método `plot()`

También podemos `posicionar` las leyendas en la mejor ubicación posible sin que interfiera con las líneas y de forma automática

```python
# Configurar etiquetas de cada línea
plt.plot(x, y, label = 'y = x**2')
plt.plot(x, y1, label = 'y = x')
# Posicionar las etiquetas en el mejor lugar
plt.legend(loc = 'best')
plt.show()
```

### Crear múltiples gráficas

##### `subplot()`

Si quisieramos graficar ambas funciones, pero en diferentes cuadrículas tenemos que emplear el método `subplot()`.

Este método se manda llamar **__después__** de un `plot()`

Al llamar al método estamos creando un 'arreglo' donde podremos ubicar cada gráfica.

**subplot**(`filas`, `columnas`, `posición de la gráfica`)

```python
plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y1, 'g')

plt.show()
```

### Modificar tamaño de las gráficas

##### figure()

Para hacer esto debemos llamar al método `figure()`, que al pasarle el parámetro `figsize = (ancho, alto)` se modificará la gráfica.

```python
# Configurar ancho y alto
plt.figure(figsize = (8,4))

plt.subplot(1, 2, 1) # 1 rows, 1 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 rows, 2 columns, 2nd subplot
plt.plot(x, y1, 'g')

plt.show()
```

### SCATTER PLOTS

##### `scatter()`

Este tipo de gráfico, a diferencia de las vistas anteriormente, están hechas de `puntos` en vez de **líneas**

```python
from numpy.random import rand

x, y = rand(2, 100)              # Array de 2 filas y 100 columnas, de valores entre 0 y 1

plt.scatter(x, y, color='red')   # Grafica las coordenadas (x, y) entre ambos arrays y de color ROJO
plt.show()
```

Si lo aplicaramos a nuestra función cuadrática, se vería lo siguiente:

```python
x = np.linspace(-1, 1, 100)
y = x**2

plt.scatter(x, y, color='green', s=2) # s configura el tamaño de los puntos (20-30 por default)
plt.show()
```

### HISTOGRAMAS

`hist()`

**Recordando:** un histograma representa la frecuencia con la que aparecen ciertos valores o rangos

La funcón recibe 3 parámetros:

**hist(data, intervalos, ancho_línea)**

* **Data:** es la lista de valores.
* **Número de intervalos:** cantidad de rangos7intervalos en los que se dividirán los datos.
* **Ancho de las barras:** el ancho de cada barra.

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]

plt.subplot(211) # Crea un arreglo 2x1 y posiciona la figura en la posicion 1 del arreglo
plt.hist(data, bins=10, rwidth=0.8, color='red')
plt.xlabel("Values", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

plt.subplot(212) # Posiciona la figura en la posicion 2 del arreglo 2x1
plt.hist(data, bins=5, rwidth=0.8, color='blue')
plt.xlabel("Values", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

plt.show()
```

### Guardar Figuras en disco

`savefig("nombre.png", transparent=True)`

Podemos guardar nuestras graficas como **png** y sin fondo.

**NOTA:** la figura se guardará en el mismo directorio de esta notebook.

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label="x**2")
plt.plot(x, y2, 'g', label="x+1")
plt.legend(loc="best")

plt.savefig("mi_grafica.png", transparent=True)
```
